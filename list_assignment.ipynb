```python
# Create a list of the first 20 positive integers. Print the list.
num_list = [i for i in range(1,21)]
print(num_list)
```

```python
# Print the first, middle, and last elements of the list created in Assignment 1.
print("First element", num_list[0])
print("Last element",num_list[-1])
print("Middle element", num_list[len(num_list)//2])
```

```python
# Print the first five elements, the last five elements, and the elements from index 5 to 15 of the list created in Assignment 1.

print("First 5 elements", num_list[:5])
print("Last 5 elements", num_list[-1:-6:-1])
print("Elements from index 5 to 15",num_list[5:16])
```

```python
# Create a new list containing only the even numbers from the list created in Assignment 1 
# using a list comprehension. Print the new list.

even_numbers = [num for num in num_list if num%2 ==0]
print(even_numbers)
```

```python
# Create two lists of the same length. Use the `zip` function to combine these lists into a list of tuples and print the result.
l1 = [1,2,3,4,5]
l2 = [56,7,8,9,10]

zipped_list = list(zip(l1,l2))
print(zipped_list)
```
